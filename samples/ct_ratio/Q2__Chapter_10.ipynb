Importing stuff:

```python
%matplotlib inline
%config InlineBackend.figure_formats = {'svg',}

import random
import numpy as np
import matplotlib.pyplot as plt
```

Create two data sets of random numbers. 10000 points between -100 and 100. 
"set_1" is uniformely distributed, "set_2" normally.

```python
set_1 = np.random.random(10000)
set_1 = set_1 * 200 - 100
set_2 = np.random.normal(0,57,10000)
```

Calculate mean and standard deviation of the two sets:

```python
print ("The mean and standard deviation of the first set are", round(np.mean(set_1),2), 
       "and", round(np.std(set_1),2),", respectively.")
print ("The mean and standard deviation of the second set are", round(np.mean(set_2),2), 
       "and", round(np.std(set_2),2),", respectively.")
```

Plot histograms of the two data sets:

```python
fig = plt.figure()
axes = fig.add_subplot(111)
axes.hist(set_1,20,(-100,100))
axes.set_title("Set 1")
axes.set_xlabel("Bins")
axes.set_ylabel("Num. occurences")
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_subplot(111)
axes.hist(set_2,40,(-200,200))
axes.set_title("Set 2")
axes.set_xlabel("Bins")
axes.set_ylabel("Num. occurences")
plt.show()
```

**Twodimensional data sets**

Create two normally distributed data sets:

```python
set_3 = np.random.normal(0,size=1000)
set_4 = set_3 + np.random.normal(0,size=1000)/2
set_5 = - set_3 + np.random.normal(0,size=1000)/2
```

Quickly check histogramms:

```python
fig = plt.figure()
axes = fig.add_subplot(111)
axes.hist(set_4,20,(-6,6))
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_subplot(111)
axes.hist(set_5,20,(-6,6))
plt.show()
```

These look very much the same. However, he then continues to plot some sort of joint scatterplot, which looks different for both.

```python
fig = plt.figure()
axes = fig.add_subplot(111)
axes.scatter(set_3,set_4, color = 'blue',label="set_4")
axes.scatter(set_3,set_5,color = 'grey',label="set_5")
axes.legend(loc=9)
axes.set_title("Very different joint distributions")
plt.show()
```

The same difference becomes apparent if one looks at the correlations of the two distributions:

```python
print (np.correlate(set_3,set_4))
print (np.correlate(set_3,set_5))
```
