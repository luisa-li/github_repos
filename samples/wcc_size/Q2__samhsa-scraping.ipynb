```python
import os
import re
import time
import requests
import datetime
import pandas as pd

from selenium import webdriver
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait

from bs4 import BeautifulSoup
from itertools import product
from string import ascii_uppercase
```

```python
## Full download of SAMHSA data:
samhsa_full = pd.read_csv("BHTFL_2017_03_12_203916.csv")
print(samhsa_full.shape)

## Source: http://www.zipcodestogo.com/Illinois/
zips = pd.read_csv("illinois-zip-codes.csv")
zips.head()
```

```python
name1_array = []
name2_array = []
address1_array = []
address2_array = []
additional_info_table_array = []
```

```python
for zip in zips["Zip Code"][5:]:
    print("ZIP ", zip)
    
    driver = webdriver.Chrome()
    driver.wait = WebDriverWait(driver, 5)
    driver.get("https://findtreatment.samhsa.gov/locator/home")
    
    search_box = driver.wait.until(EC.presence_of_element_located((By.ID, "sAddr")))
    search_box.send_keys(zip)

    go_button = driver.wait.until(EC.presence_of_element_located((By.ID, "submit")))
    go_button.click()

    driver.implicitly_wait(10) 
    data_table = driver.wait.until(EC.presence_of_element_located((By.ID, "data-table-facility")))
    
    for i in range(0,50):
        
        rows = driver.find_elements_by_class_name("main-row")

        print("Row #: ", i)
        row = rows[i]
        print(row.text)
        
        # info = driver.wait.until(EC.presence_of_element_located((By.CLASS_NAME, "listing-col2")))
        
        info = row.find_elements_by_class_name("listing-col2")
        divs = info[0].find_elements_by_tag_name("div")
        name1 = divs[0].text
        name2 = divs[1].text
        address1 = divs[2].text
        address2 = divs[3].text

        more_info = row.find_elements_by_class_name('listing-col3')

        href = more_info[0].find_elements_by_class_name('more-information')
        href = href[0]

        driver.execute_script("return arguments[0].scrollIntoView();", href)
        href.click()
        time.sleep(2)

        more_table = driver.wait.until(EC.presence_of_element_located((By.CLASS_NAME, "facility-more-info")))

        name1_array.append(name1)
        name2_array.append(name2)
        address1_array.append(address1)
        address2_array.append(address2)
        additional_info_table_array.append(more_table.text)
        
        time.sleep(5)
    
    driver.close()
    driver.quit()
```
